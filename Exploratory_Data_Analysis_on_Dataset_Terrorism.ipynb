**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

**Import Dataset**

```python
data=pd.read_csv('/content/drive/MyDrive/LGM VIP TASK 4/Global Terrorism - START data.zip',encoding='latin-1')
```

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
```

```python
data['Casualities'] = data['kill'] + data['Wound']
```

```python
data.info()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count)
plt.xticks(rotation = 50)
plt.xlabel('Year of Terrorist Attacks',fontsize=25)
plt.ylabel('Number of Terrorist Attacks',fontsize=25)
plt.title('Year of Terrorist Attacks',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Regional Terrorist Activities in Each Year',fontsize=30)
plt.ylabel('Number of Terrorist Attacks',fontsize=25)
plt.xlabel("Year of Terrorist Attacks",fontsize=25)
plt.show()
```

```python
attack = data.Country.value_counts()[:10]
attack
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values)
plt.title('Countries Which Are Affected',fontsize=30)
plt.xlabel('Countries',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.xticks(rotation = 50)
plt.show()
```

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.8,ax=ax4)
plt.xticks(rotation = 50)
plt.title("Deaths on Attack",fontsize=30)
plt.ylabel("Number of People Died",fontsize=25)
plt.xlabel('Year',fontsize=25)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10))
plt.xticks(rotation=50)
plt.title("Number of People Killed",fontsize=30)
plt.ylabel('Count of people',fontsize=25)
plt.xlabel('Type of Attack',fontsize=25)
plt.show()
```

```python
data['Attacktype'].value_counts().plot(kind='bar',figsize=(10,10),color='pink')
plt.xlabel("Type of Attack",fontsize=25)
plt.ylabel("Count of Attack",fontsize=25)
plt.title("Attacking",fontsize=30)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',figsize=(10,10))
plt.title("Terrorist Group Attacks",fontsize=30)
plt.xlabel("Name of the Terrorist Group",fontsize=25)
plt.ylabel("Number of People Killed",fontsize=25)
plt.show()
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(10,10))
plt.xticks(rotation = 50)
plt.xlabel("Cities",fontsize=30)
plt.ylabel("Count of attack",fontsize=30)
plt.title("Most Affected Cities",fontsize=35)
plt.show()
```
